```python
plt.figure(figsize=(20,40))
sns.countplot(y="CITY",data=df)
plt.title("City-wise Sales")
plt.show()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
impute_mode=SimpleImputer(strategy='most_frequent')
impute_mode.fit(df[["STATE","POSTALCODE","TERRITORY"]])

df[["STATE","POSTALCODE","TERRITORY"]]=impute_mode.transform(df[["STATE","POSTALCODE","TERRITORY"]])
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['MSRP'])
```

```python
df['MSRP']=np.where(df['MSRP']>205.0,205.0,df["MSRP"])
```

```python
sns.boxplot(df['MSRP'])
```

```python
#We have removed the outlier
```

```python
df.corr()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df['YEAR_ID'].value_counts().plot(kind="pie",figsize=(6,6),autopct="%.2f%%")
plt.title('Year')
plt.show()
```

```python
# From the pie chart it is clear that the 2004 had the maximum number of sales, 2003 had the second highes sales
# and 2005 had the lowest number of sales.
```

```python
sns.distplot(df['PRICEEACH'],color='green')
plt.title('Price of Products')
```

```python
plt.scatter(x=df['SALES'],y=df['QUANTITYORDERED'])
plt.title('Sales vs Quantity Ordered')
```

```python
# From the Scatter plot we can say that the maximum Quantity Ordered is in the range from 20 to 50 for Sales in the range
# around 500 to approx. 8000.
```

```python
sns.pairplot(df,height=2)
```

```python
sns.swarmplot(x="YEAR_ID",y="SALES",data=df)
plt.title("Year-wise Sales")
```

```python
# From the above Swarm plot we can conclude that
# 1. for the year 2003 the Sales happened from approx. 1000 that went upto approx. 6000.
# 2. for the year 2004 the Sales happened from approx. 1000 that went upto approx. 7000.
# 3. for the year 2005 the Sales happened from approx. below 1000 that went upto approx. 5000.
```

```python
sns.barplot(x='DEALSIZE',y="SALES",data=df)
plt.title('Dealsize-wise Sales')
```

```python
# From the above bar plot it is clear that Large deal size have highest Sales of more than 8000.
# Medium deal size have Sales of more than 4000 and Small deal size have Sales of around 2000.
```
