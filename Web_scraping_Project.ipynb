```python
import requests
```

```python
site = requests.get("https://www.flipkart.com/search?q=samsung%20galaxy%20m%20series&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
site
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(site.content,"html.parser")
```

```python
title = soup.find("a")
title
```

```python
title = soup.find_all("div")
title
```

```python
price = soup.find_all("div",attrs = {"class":"_3tbKJL"})
price
```

```python
name = soup.find_all("div",attrs = {"class":"_4rR01T"})
name
```

```python
details = soup.find_all("div",attrs = {"class":"fMghEO"})
details
```

```python
for i in price:
  name.append(i.text)
```

```python
import pandas as pd
df=pd.DataFrame({"Model":name[0:15],"Price":price[0:15],"Specifications":details[0:15]})
df
```
