```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#For project the file chosen is sales data of popular mobile phones brand of India from Flipkart.
```

```python
df=pd.read_csv("Flipkart Mobile - 2.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["brand"].hist()
plt.title("Brands")
plt.show()
```

```python
#The above histogram shows that the maximum number of mobiles sold are of Realme brand, follwed by Samsung, Xiaomi.
#The sales of Poco & Apple brands are almost equal.
```

```python
plt.figure(figsize=(20,30))
sns.countplot(data=df,y="model")
plt.title("Model")
plt.show()
```

```python
#The above graph shows the highest selling model which is IPhone XR, followed by IPhone 12 & IPhone 12 mini.
```

```python
df["model"].value_counts()
```

```python
plt.figure(figsize=(20,30))
sns.countplot(data=df,y="sales_price")
plt.title("Price of Mobiles")
plt.show()
```

```python
#The above graph shows the sales price of mobile phones. From the graph we know that the price category for maximum number of 
#mobile phones is Rs 14999, followed by Rs 15999.
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df,x="brand",hue="processor")
plt.title("Processors")
plt.show()
```

```python
#The graph above shows different processors used by brands for their mobile phones.
#From the graph we can conclude that Realme phones have Media Tek processors significantly, Apple has maximum ceramic processors
#for their phones, Poco uses only two processors viz., maximum Qualcomm and Media Tek, Samsung uses Exynos processors maximum
#for their mobiles and Xiaomi use Qualcomm processors mostly.
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df,x="brand",y="sales_price")
plt.title("Price")
plt.show()
```

```python
#The above boxplot gives the information of price range of each brand.
#Apple has priced its phones starting from approx. Rs 35000 to maximum Rs. 80000.
#Poco has a price range for its phones from Rs. 10000 to maximum Rs 30000.
#Realme price range for its mobiles is approx Rs.7000 and maximum Rs 40000.
#Samsung has a priced its mobile starting from Rs 9000 and its top model approx. Rs 160000.
#Xiaomi has mobiles starting from approx. Rs 6000 to Rs. 55000.
```

```python
plt.subplot(2,2,2)
df["num_rear_camera"].value_counts().plot(figsize=(8,8),kind="bar",color="black")
plt.title("Number of rear cameras")
plt.show()
```

```python
#The above graph shows that maximum number of mobile phones will have 3 rear cameras rather than 4 rear cameras.
```

```python
df.corr()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df["RAM"].value_counts().plot(kind="pie",figsize=(6,6),autopct='%.2f%%')
plt.title("Type of RAM")
plt.show()
```

```python
#The pie chart above shows that most number of mobiles will have 4GB of RAM followed by 6GB of RAM and 8GB of RAM.
```

```python
df.groupby("brand")["battery_capacity"].mean().plot(kind="bar")
plt.title("Battery Capacity")
plt.show()
```

```python
#The above graph shows the battery capacity of mobiles for each brand.
#Poco mobiles have the highest battery capacity of more than 5000mAh.
#Realme and Samsung mobiles have the same battery capacity of around 4800mAh.
#Xiaomi mobiles have a battery capacity of around 4000mAh.
#Apples mobiles have a battery capacity of around 3800mAh.
```

```python
me=df["display_size"].mean()
med=df["display_size"].median()
print(me)
print(med)
```

```python
sns.distplot(df["display_size"])
plt.axvline(me,color="g")
plt.axvline(med,color="r")
plt.title("Screen Size")
plt.show()
```

```python
#As seen in the above plotted graph of various display or screen size.
#From the graph we know that maximum number of mobiles have display size of 6.5 inches.
```

```python
sns.pairplot(df[["brand","RAM","ROM"]],hue="brand")
plt.show()
```

```python
#From the top pair plots we can conclude the following
#Apple phones have ROM starting from 50GB to 250GB and RAM from 3GB to 4GB.
#Poco phones have ROM starting from 30GB to 250GB and RAM from 4GB to 8GB.
#Realme phones have ROM starting from 30GB to 250GB and RAM from 4GB to 8GB.
#Samsung phones have ROM starting from 30GB to 500GB and RAM from 4GB to 12GB.
#Xiaomi phones have ROM starting from 30GB to 250GB and RAM from 4GB to 8GB.
```

```python
sns.pairplot(df[["sales_price","sales","brand"]],hue="brand")
plt.show()
```

```python
#From the above pair plots we can conclude the following
#Apple mobiles have a sales of approx. 150 mobiles.
#Poco mobiles have sales of approx. 420 mobiles.
#Realme mobiles have sales of nearly 180 mobiles.
#Samsung mobiles have a sales of nearly 100 mobiles.
#Xiaomi mobiles have sales of approx 100 mobiles.
```
